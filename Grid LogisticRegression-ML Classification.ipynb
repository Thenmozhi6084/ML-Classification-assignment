```python
from pandas import Series, DataFrame
dataset1={'age','bp','al','su','bgr','bu','sc','sod','pot','hrmo','pcv','wc','rc'}
dataset2={
    'rbc':['abnormal','normal'],
    'pc':['abnormal','normal'],
    'pcc':['notpresent','present'],
    'ba':['notpresent','present'],
    'htn':['no','yes'],
    'dm':['no','yes'],
    'cad':['no','yes'],
    'appet':['poor','yes'],
    'pe':['poor','yes'],
    'ane':['no','yes']
}
df1=pd.DataFrame(dataset2)
dummies1=pd.get_dummies(df1,prefix=['rbc','pc','pcc','ba','htn','dm','cad','appet','pe','ane'])
dataset3={'sg':['a','b','c','d']}
df2=pd.DataFrame(dataset3)
dummies2=pd.get_dummies(df2,prefix=['sg'])
print(dummies1)
print(dummies2)
```

```python
df=pd.concat([dataset1,dummies1.reset_index(drop=True),dummies2.reset_index(drop=True)],axis='columns',ignore_index=True)
print(df)
```
